```python
tweet_count = 15000
text_query = "iphone 14"
since_date = "2022-09-17"
until_date = "2022-12-11"

os.system('snscrape --jsonl --max-results {} --since {} twitter-search "{} until:{}"> iphone_14_post_release_tweets.json'.format(tweet_count, since_date, text_query, until_date))
```

```python
tweets_df1 = pd.read_json('iphone_14_post_release_tweets.json', lines=True)
tweets_df1
```

```python
tweets_df1.to_csv('iphone_14_post_release.csv', sep=',', index=False)
```
