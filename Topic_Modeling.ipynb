## iPhone 13 - pre release

```python
iphone_13_pre = pd.read_csv('/content/drive/My Drive/Data Mining/iphone_13_pre_release.csv')
print("Dataframe shape:", iphone_13_pre.shape)
iphone_13_pre.head()
```

```python
df1 = iphone_13_pre.loc[(iphone_13_pre.lang == "en"),:]
df1['lang'].value_counts()
```

```python
# Remove the columns
TopicModelling_Data = df1.drop(columns=['_type', 'url', 'user','replyCount','retweetCount','likeCount','quoteCount','conversationId','lang','source','sourceUrl','sourceLabel','media','retweetedTweet','quotedTweet','inReplyToTweetId','inReplyToUser','mentionedUsers','coordinates','place','hashtags','cashtags','card'], axis=1).sample(100)
# Print out the first rows of papers
TopicModelling_Data.head()
```

```python
# Load the regular expression library
import re
# Remove punctuation
TopicModelling_Data['rawContent'] = \
TopicModelling_Data['rawContent'].map(lambda x: re.sub('[,\.!?]', '', x))
# Convert the titles to lowercase
TopicModelling_Data['rawContent'] = \
TopicModelling_Data['rawContent'].map(lambda x: x.lower())
# Print out the first rows of papers
TopicModelling_Data['rawContent'].head()
```

```python
import gensim
from gensim.utils import simple_preprocess
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
stop_words.extend(['from','http','https','スマートフォン','g対応','simフリー','ريال'])
def sent_to_words(sentences):
    for sentence in sentences:
        # deacc=True removes punctuations
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) 
             if word not in stop_words] for doc in texts]


data = TopicModelling_Data.rawContent.tolist()
data_words = list(sent_to_words(data))
# remove stop words
data_words = remove_stopwords(data_words)
print(data_words[:1][0][:30])
```

```python
import gensim.corpora as corpora
# Create Dictionary
id2word = corpora.Dictionary(data_words)
# Create Corpus
texts = data_words
# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]
# View
print(corpus[:1][0][:30])
```

```python
from pprint import pprint
# number of topics
num_topics = 10
# Build LDA model
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2word,
                                       num_topics=num_topics)
# Print the Keyword in the 10 topics
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```

```python
import pyLDAvis
import pyLDAvis.gensim_models
import pickle 
# Visualize the topics
pyLDAvis.enable_notebook()
LDAvis_data_filepath = os.path.join('/content/drive/My Drive/Data Mining/13_pre'+str(num_topics))
# # this is a bit time consuming - make the if statement True
# # if you want to execute visualization prep yourself
if 1 == 1:
    LDAvis_prepared = pyLDAvis.gensim_models.prepare(lda_model, corpus, id2word)
    with open(LDAvis_data_filepath, 'wb') as f:
        pickle.dump(LDAvis_prepared, f)
# load the pre-prepared pyLDAvis data from disk
with open(LDAvis_data_filepath, 'rb') as f:
    LDAvis_prepared = pickle.load(f)
pyLDAvis.save_html(LDAvis_prepared, '/content/drive/My Drive/Data Mining/13_pre'+ str(num_topics) +'.html')
LDAvis_prepared
```
