```python
tweet_count = 5000
text_query = "iphone 13"
since_date = "2021-09-22"
until_date = "2022-09-22"

os.system('snscrape --jsonl --max-results {} --since {} twitter-search "{} until:{}"> iphone_13_pre_release_tweets.json'.format(tweet_count, since_date, text_query, until_date))
```

```python
tweets_df1 = pd.read_json('iphone_13_pre_release_tweets.json', lines=True)
tweets_df1
```

```python
tweets_df1.to_csv('iphone_13_pre_release.csv', sep=',', index=False)
```
